```python
import pandas as pd
bid = pd.read_csv("/prj/Qingyang/BID_seq_HEK293T.bed", sep='\t', low_memory=False)
psi = pd.read_csv("/prj/Qingyang/psi_cov10.bed", sep='\t', low_memory=False)
```

```python
bid['site'] = bid['chrom'] + ':' + bid['chromStart'].astype(str) + '-' + bid['chromEnd'].astype(str)
psi['site'] = psi['chrom'] + ':' + psi['chromStart'].astype(str) + '-' + psi['chromEnd'].astype(str)
```

```python
bid_threshold = 50
psi_threshold = 50

bid_set = set(bid[bid['score'] >= bid_threshold]['site'])
psi_set = set(psi[psi['frequency'] >= psi_threshold]['site'])
```
