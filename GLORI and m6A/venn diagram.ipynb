```python
import pandas as pd
glori = pd.read_csv("/prj/Qingyang/GLORI.chrALL.tsv", sep='\t', low_memory=False)
m6a = pd.read_csv("/prj/Qingyang/m6A_cov10.bed", sep='\t', low_memory=False)
```

```python
glori['site'] = glori['chrom'] + ':' + glori['chromStart'].astype(str) + '-' + glori['chromEnd'].astype(str)
m6a['site'] = m6a['chrom'] + ':' + m6a['chromStart'].astype(str) + '-' + m6a['chromEnd'].astype(str)
```

```python
glori_threshold = 50
m6a_threshold = 50

glori_set = set(glori[glori['score'] >= glori_threshold]['site'])
m6a_set = set(m6a[m6a['frequency'] >= m6a_threshold]['site'])
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2

plt.figure(figsize=(6,6))
plt.title("GLORI vs. m6A Venn Diagram", fontsize=14)
venn2([glori_set, m6a_set], set_labels=('GLORI', 'm6A'))
for label in venn.set_labels:
    label.set_fontsize(14) 
for label in venn.subset_labels:
    label.set_fontsize(16)
plt.show()
```
